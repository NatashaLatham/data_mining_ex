
1.2.2 Principal Component Analysis, more commonly known as PCA, can be cleverly used to better visualize high dimensional data. 

> a. (1 point) Explain what PCA is and when it can be used. Make sure to provide an in-depth explanation and note what the drawbacks and limitations are.


Principal components analysis (PCA) is a linear matrix factorization technique which can be applied to square data matrices (In the case of non-square data you can use the related concept of SVD, or apply PCA to the correlation/covariance matrices of the data). 

Applying PCA can be thought of as a change of basis vectors for describing the data, such that the direction of these basis vectors coincides with the axes of greatest variability in the data. Importantly, these new basis vectors are orthogonal to one another. The first PCA vector lies along the axis of greatest variability, and consecutive vectors lie along the axis of greatest -remaining- variability, provided that it is orthogonal to all preceding vectors. This enables PCA to identify the strongest patterns in the data. You will find a number of components equal to the rank of the original matrix to which the PCA is applied. The total variability is preserved, but the attributes are now uncorrelated; the new attributes will have a covariance of 0 between each pair. Each PC vector has an accompanying value which describes the amount of variance that is explained by that particular PC.

Often, the first few principal components capture a substantial amount of the variability in the data. In these cases, it can be advantageous to continue the analysis only with the top principal components. This results in lower-dimensional data, which might be more amenable to further analysis techniques, and might be less noisy.



Limitations:

1. PCA is only suitable for uncovering hidden LINEAR relationships in the data. If there are non-linear relationships in the data (eg it is shaped like a spiral), then PCA is not very informative.

2. If you do not normalize the attributes, the relative scale of the attributes can affect the direction of the PC's.

3. Time/space complexity of PCA increases with the square of the amount of objects in the dataset. This can make it impractical to apply on very large datasets.








To apply PCA we must first mean center the data. mean centering means that the mean value for an attribute is subtracted from that attribute. 

> b. (0.5 points) Why do we first need to mean center the data before applying PCA?

Because PC vectors start in the origin, in this case your components will also be describing the mean of the data. Especially if the means of the attributes is high relative to their variance, your first PC will be pointing towards the mean. This also makes your other PCs less informative as they are constrained to be orthogonal to the first (uninformative) PC. 

When using PCA, you are generally interested in the -relationship- among datapoints themselves, and not their mean.

In practice, applying PCA comes down to a series of matrix operations, this is useful, as matrix operations can be applied with little effort and can be computed in relatively little time.

The given data **X**, must first be mean centered. This can be easily done by calculating a row vector **μ** containing the mean values of each attribute and then subtracting it from **X** (**Y** = **X** - **1μ**, where **1** is a column vector of size **N** x 1, with N indicating the number of observations in the original data).

Then, the Singular Value Decomposition, or SVD, of **Y**, the mean centered data, can be calculated. **Y** = **USV**<sup>T</sup>. In practice, this is often done using the numpy.linalg.svd() function.

Using SVD on **Y** yields a series of vectors which can be used to project the data onto specific Principal Components (PCs). The entire dataset can be projected onto the Principal Components by multiplying **Z** = **Y\*V**, where **Z** indicates the projected data. Alternatively, one could project onto just a subset of all the PCs by indexing in the multiplication. For example: **Z** = **Y\*V[:,1]** would yield the projection of the data onto the first PC.

> d. (1 point) Alternatively, one could use EigenValue Decomposition, EVD, instead of SVD. What are the similarities and differences between SVD and EVD? Can both be applied in each case?

Eigendecomposition: A= P*D*inv(P)
SVD:                A= U*S*V

We cannot apply both techniques in every case where we want to analyze a matrix. Eigendecomposition requires the matrix in question to be both square and symmetrical.




Similarities

1. They both factorize a matrix into three matrices, the middle of which is a diagnonal matrix.

2. They can both be used to find patterns of main variation in a matrix, and these patterns can inform dimenionality reductions strategies.


Differences

1. vectors in the eigendecompositionmatrix P are not necessarily orthonormal,  whereas vectors in the SVD matrices are orthonormal.

2. In eigendecomposition, the nondiagonal matrices P and inv(P) are each others inverse, whereas for SVD the nondiagonal matrices U and V may be very different from one another, and depend on the dimensions of the initial matrix.

3. SVD can always be performed for any rectangular matrix, whereas eigendecomposition only works for some symmetrical square matrices.







PCA attempts to explain as much of the variance in data with as few PCs as possible. The variance explained by each of the PCs **m** can be calculated using the following formula: 
$\rho_m = 1 - \frac{\|{Y} - {u}_m s_{mm} {v}_m^T\|^2_F}{\|{Y}\|^2_F} = \frac{s_{mm}^2}{\displaystyle \sum_{m'=1}^M s^2_{m'm'}}$

Which indicates that the variation $\rho$ for a given **m** can be calculated by dividing the squared singular value of component **m** by the sum of all squared singular values. 
> e. (1 point) Calculate the row vector $\rho$ containing all values of $\rho$ for all PCs **m**. Create a bar plot with the variance explained on the Y axis and the number of the PC on the X axis. 

```python
S_squared=np.power(S,2)
S_squared_sum=np.sum(S_squared)
S_rho=S_squared/S_squared_sum

# The row vector with all values of rho for al PCs.
print(S_rho)


plt.bar(range(8),S_rho)
plt.xlabel('Principal component')
plt.ylabel('Variance explained')
plt.show()
print('Amount of variance explained per PC from the nanonose data')
```

## 1.3 Similarity measures

We will use a subset of the data on wild faces described by Berg in 2005 transformed to a total
of 1000 gray scale images of size 40x40 pixels, we will attempt to find faces in the
data base that are the most similar to a given query face. To measure similarity we
will consider the following measures: SMC, Jaccard, Cosine, ExtendedJaccard, and
Correlation. These measures of similarity are described in *Introduction to Data Mining*, page 73-77 and are given by

\begin{eqnarray*}
\textrm{SMC}({x},{y}) & = & \frac{\textrm{Number of matching attribute values}}{\textrm{Number of attributes}} \\
\textrm{Jaccard}({x},{y}) & = & \frac{\textrm{Number of matching presences}}{\textrm{Number of attributes not involved in 00 matches}} \\
\textrm{Cosine}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\| \|{y}\|} \\
\textrm{ExtendedJaccard}({x},{y}) & = & \frac{{x}^T {y}}{\|{x}\|^2 + \|{y}\|^2 - {x}^T {y}} \\
\textrm{Correlation}({x},{y}) & = & \frac{\textrm{cov}({x},{y})}{\textrm{std}({x}) \textrm{std}({y})}
\end{eqnarray*}
where $\textrm{cov}({x},{y})$ denotes the covariance between ${x}$ and ${y}$ and $\textrm{std}({x})$ denotes the standard deviation of ${x}$.

Notice that the SMC and Jaccard similarity measures only are defined for binary
data, i.e., data that takes values in $\{0,1\}$. As the data we analyze is non-binary,
we will transform the data to be binary when calculating these two measures of
similarity by setting
$x_i = \left\{ \begin{array}{ll} 0 & \textrm{if~} x_i < \textrm{median}({x}) \\
                                1 & \textrm{otherwise.} \end{array} \right.$
