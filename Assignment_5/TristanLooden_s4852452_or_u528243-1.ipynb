If we assume that $c_1$, $c_2$, and $c_3$ have different values, what are the possible inequality relationships (e.g. $c_1 \leq c_2 \leq c_3$) among $c_1$, $c_2$, and $c_3$? 2. (0.5 points) Suppose the confidence of the rules  $ \text{ {a } } \rightarrow   \text{{b }} $ and $ \text{ {b } } \rightarrow   \text{{c }} $ are larger than the confidence threshold. Is it possible that $ \text{ {a } } \rightarrow   \text{{c }} $ has a confidence below that threshold? If no, explain why. If yes, give an example. 

QUESTION 1

From theorem 6.2 in the book we know that if rule X -> Y - X, it is the case that for any subset X' of X, it is the case that X' -> Y - X' must necessarily have a lower (or equal) confidence. Now we can first of all assert that it must be the case that c2 $\leq$  c3. for c1, this is not as trivial as it is not the same set as c2 and c3. However, as confidence is computed as the support of the whole set divided by the support of the antecedent, and c1, c2 have the same antecedent, and we know from the antimonotonicity property of itemset support that support for {a,b} must be greater or equal to support for {a,b,c}. This means that support{a,b}/support{a} will be greater or equal to support{a,b,c}/support{a}. This means that confidence c2 $\leq$ c1. 

There now still are two possible relationships between the three values: c2 $\leq$ c3 $\leq$ c1 or c2 $\leq$ c1 $\leq$ c3. We can however be certain that c2 has the lowest confidence of the three.



QUESTION 2

This is certainly possible. Take, for instance the following dataset:

1. a,b
2. b,c
3. a,b
4. a,b
5. a,b
6. b,c
7. a,c
8. b,c

The confidence for rule a -> b is 4/5 = 80% 
The confidence for rule b -> c is 3/3 = 100%
However, the confidence for rule a -> c is only 1/5 = 20%

Therefore we see that while rule a -> b and b -> c may be above some threshold (in this example, it could be anywhere between 21% and 80%), rule a -> c can be below it.
